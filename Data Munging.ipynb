```python
filepath = os.path.join('winemag-data-130k-v2.csv')
master_df = pd.read_csv(filepath)
master_df.head()
```

```python
master_df = master_df[['title', 'variety', 'winery', 'country', 'province', 'region_1', 'region_2', 'description', 'designation', 'points', 'price', 'taster_name', 'taster_twitter_handle']]
master_df.head()
```

```python
winery_df = master_df[['winery', 'country', 'province']]
winery_df = winery_df.drop_duplicates().reset_index()[['winery', 'country', 'province']]
winery_df.head()
```

```python
province_df = master_df[['province', 'country']]
province_df = province_df.drop_duplicates().reset_index()[['province', 'country']]
province_df.head()
```

```python
country_df = master_df[['country']]
country_df = country_df.drop_duplicates().reset_index()[['country']]
country_df.head()
```

```python
taster_df = master_df[['taster_twitter_handle', 'taster_name']]
taster_df = taster_df.drop_duplicates().reset_index()[['taster_twitter_handle', 'taster_name']]
taster_df.head()
```

```python
taster_list = []
for index, row in master_df.iterrows():
    taster = row['taster_name']
    handle = row['taster_twitter_handle']
    for index, row in taster_df[taster_df.taster_name == taster].iterrows():
        if taster == row['taster_name'] and handle == row['taster_twitter_handle']:
            taster_id = int(index)
        else:
            taster_id = None
    taster_list.append(taster_id)
```

```python
master_df = master_df.drop(columns = 'taster_id')
master_df['taster_id'] = taster_list
master_df.head()
```
