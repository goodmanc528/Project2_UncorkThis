```python
import pandas as pd
import os
from config import gkey
from config import gpkey
import requests
import json
```

```python
filepath = os.path.join('country_centroids_all.csv')
coordinates_df = pd.read_csv(filepath, delimiter = '\t')
coordinates_df.head()
```

```python
filepath = os.path.join('winemag-data-130k-v2.csv')
master_df = pd.read_csv(filepath)
master_df.head()
```

```python
master_df = master_df[['title', 'variety', 'winery', 'country', 'province', 'region_1', 'region_2', 'description', 'designation', 'points', 'price', 'taster_name', 'taster_twitter_handle']]
master_df.head()
```

```python
country_df = master_df[['country']]
country_df = country_df.drop_duplicates().reset_index()[['country']]
country_df.head()
```

```python
lat = []
lon = []
for index, row in country_df.iterrows():
    country = row['country']
    if country == 'US':
        country = 'United States'
    if country == 'England':
        country = 'United Kingdom'
    temp_df = coordinates_df[coordinates_df.SHORT_NAME == country]
    if len(temp_df) == 1:
        lat.append(temp_df.iat[0,0])
        lon.append(temp_df.iat[0,1])
    else:
        lat.append(None)
        lon.append(None)
country_df['lat'] = lat
country_df['lon'] = lon
country_df.head()
```

```python
province_df = master_df[['province', 'country']]
province_df = province_df.drop_duplicates().reset_index()[['province', 'country']]
province_df.head()
```

```python
len(province_df)
```

```python
latitude = []
longitude = []
i = 0
for index, row in province_df.iterrows():
    province = row['province']
    country = row['country']
    place = f'{province}, {country}'
    mapUrl = 'https://maps.googleapis.com/maps/api/place/findplacefromtext/json?inputtype=textquery'
    fields = 'id,geometry'
    url = f'{mapUrl}&input={province}&fields={fields}&key={gkey}'
    data = requests.get(url).json()
    try:
        lat = data['candidates'][0]['geometry']['location']['lat']
        lon = data['candidates'][0]['geometry']['location']['lng']
        latitude.append(lat)
        longitude.append(lon)
    except:
        temp_df = country_df[country_df.country == country]
        if len(temp_df) > 0:
            lat = temp_df.iat[0,1]
            lon = temp_df.iat[0,2]
            latitude.append(lat)
            longitude.append(lon)
        else:
            lat = None
            lon = None
            latitude.append(lat)
            longitude.append(lon)
    i += 1
    if i % 25 == 0:
        print(i)
province_df['pro_lat'] = latitude
province_df['pro_lon'] = longitude
province_df.head()
```

```python
country_ids = []
for index, row in province_df.iterrows():
    country = row['country']
    temp_df = country_df[country_df.country == country]
    try:
        country_id = temp_df.reset_index().iat[0,0]
        country_ids.append(country_id)
    except:
        country_ids.append(None)
province_df['country_id'] = country_ids
province_df.head()
```

```python
province_df = province_df.drop(columns = 'country')
province_df.head()
```

```python
winery_df = master_df[['winery', 'country', 'province']]
winery_df = winery_df.drop_duplicates().reset_index()[['winery', 'country', 'province']]
winery_df.head()
```

```python
country_ids = []
province_ids = []
for index, row in winery_df.iterrows():
    country = row['country']
    temp_df = country_df[country_df.country == country]
    try:
        country_id = temp_df.reset_index().iat[0,0]
        country_ids.append(country_id)
    except:
        country_ids.append(None)
    province = row['province']
    temp_pro_df = province_df[province_df.province == province]
    try:
        province_id = temp_pro_df.reset_index().iat[0,0]
        province_ids.append(province_id)
    except:
        province_ids.append(None)
winery_df['country_id'] = country_ids
winery_df['province_id'] = province_ids
winery_df.head()
```

```python
winery_df = winery_df.drop(columns = ['country', 'province'])
winery_df.head()
```

```python
taster_df = master_df[['taster_twitter_handle', 'taster_name']]
taster_df = taster_df.drop_duplicates().reset_index()[['taster_twitter_handle', 'taster_name']]
taster_df.head()
```

```python
wine_df = master_df[['title', 'variety', 'winery', 'country', 'province', 'description', 'designation', 'points', 'taster_twitter_handle']]
wine_df.head()
```

```python
country_ids = []
province_ids = []
winery_ids = []
taster_ids = []
for index, row in wine_df.iterrows():
    country = row['country']
    temp_df = country_df[country_df.country == country]
    try:
        country_id = temp_df.reset_index().iat[0,0]
        country_ids.append(country_id)
    except:
        country_ids.append(None)
    province = row['province']
    temp_pro_df = province_df[province_df.province == province]
    try:
        province_id = temp_pro_df.reset_index().iat[0,0]
        province_ids.append(province_id)
    except:
        province_ids.append(None)
    winery = row['winery']
    temp_winery_df = winery_df[winery_df.winery == winery]
    try:
        winery_id = temp_pro_df.reset_index().iat[0,0]
        winery_ids.append(province_id)
    except:
        winery_ids.append(None)
        
    taster = row['taster_twitter_handle']
    temp_taster_df = taster_df[taster_df.taster_twitter_handle == taster]
    try:
        taster_id = temp_pro_df.reset_index().iat[0,0]
        taster_ids.append(province_id)
    except:
        taster_ids.append(None)
wine_df['country_id'] = country_ids
wine_df['province_id'] = province_ids
wine_df['winery_id'] = winery_ids
wine_df['taster_id'] = taster_ids
wine_df.head()
```

```python
wine_df = wine_df.drop(columns = ['country', 'province', 'winery', 'taster_twitter_handle'])
wine_df.head()
```
